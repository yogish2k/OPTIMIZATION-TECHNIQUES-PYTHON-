## Interval Halving Method

```python
import math as mp

a = int(input("Lower Bound A value : "))
b = int(input("Upper Bound B value : "))
L = b - a
xm = (a+b)/2 # Middle point of search space
eps = float(input("Choose Epsilon value : "))
Iter = 1
#Nu_Iter = int(input("Enter the number of iterations : "))
# Function to hold the equation
def equation(x):
    return 2*x**3-96*x+128 # Change the equation as per your need


while (abs(L)>eps):
    print('\n'+"Interation - {}".format(Iter)+'\n')
    print("a = {}".format(a),"b = {}".format(b))
    print("L = {}".format(L))
    
    xm = (a+b)/2
    fxm = equation(xm)
    
    print("xm = {}".format(xm)+"\n", "f(xm) = {}".format(fxm))
    
    
    x1 = a + (L/4)
    x2 = b - (L/4)
    
    fx1 = equation(x1)
    fx2 = equation(x2)
    
    print("x1 = {}".format(x1)+'\n', "f(x1) = {}".format(fx1))
    print("x2 = {}".format(x2)+'\n', "f(x2) = {}".format(fx2))
    
    if fx1 < fxm:
        b = xm
        xm = x1
        L = b - a
        Iter += 1
    elif fx2 < fxm:
        a = xm
        xm = x2
        L = b - a
        Iter += 1
    else:
        a = x1
        b = x2
        L = b - a
        Iter += 1

print('\n'+"the solution lies b/w",a,"and ", b)
```
