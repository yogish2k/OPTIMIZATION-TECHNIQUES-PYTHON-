```python
from math import sqrt, exp, sin, cos
from sympy import *
import numpy as np
from sympy.solvers import solve

dimension=int(input('enter dim'))
a=np.arange(dimension)
a=a.astype(float)

a = [float(i) for i in input("Enter the initial values of a: ").split()]

b=np.arange(dimension)
b=b.astype(float)
b = [float(i) for i in input("Enter the final values of b: ").split()]
print(b)

func = input("Enter the function")

length=dimension
t,x,y,z,d,w,k,l,m,n= symbols('t x y z d w k l m n')

s = []

p = range(length)
for i in p:
    s.append(a[i]+t*b[i])
print(s)

fn = sympify(func)

if length==1:
     expr=fn.subs([(x, s[0])])
elif length==2:
    expr=fn.subs([(x, s[0]), (y, s[1])])
    print('ok')
elif length==3:
    expr=fn.subs([(x, s[0]), (y, s[1]), (z, s[2])])
elif length==4:
    expr=fn.subs([(x, s[0]), (y, s[1]), (z, s[2]),(w, s[3])])
    print('ok')
elif length==5:
    expr=fn.subs([(x, s[0]), (y, s[1]), (z, s[2]),(w, s[3]),(d, s[4])])


print(expr)
expr_differential=expr.diff(t)
print(expr_differential)
critical_point=solve(expr_differential,t)
print(critical_point)

print(len(critical_point))
critical=[]
double_differential=expr_differential.diff(t)
print(double_differential)
u=range(len(critical_point))
for i in u:
    g=double_differential.subs([(t, critical_point[i])])
    if g>0:
        critical.append(critical_point[i])



q=[]
for i in p:
    q.append(a[i]+((critical_point[0])*b[i]))
print(q)


```
